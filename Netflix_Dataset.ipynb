#Netflix_Dataset :

```python
import pandas as pd
import numpy as np
```

```python
var = pd.read_csv("netflix_titles.csv")
df = pd.DataFrame(var)
df
```

```python
df.columns
print()
df.shape
print()
df.info()
print()
df.describe()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"] , format = "%B %d, %Y" , errors="coerce")
```

```python
df['duration_num'] = df['duration'].str.extract('(\d+)')
df['duration_num'] = df['duration_num'].astype(float)
```

```python
df['listed_in'] = df['listed_in'].str.split(', ')
df["cast"] = df["cast"].str.split(', ')
df["country"] = df["country"].str.split(', ')
df["rating"] = df["rating"].str.split(', ')
```

```python
df.columns
print()
df.shape
print()
df.info()
print()
df.describe()
```

```python
df
```

```python
missing_values = df.isnull().sum() 
missing_values
```

```python
df = df.dropna()
df
```

```python
count_of_M_Tv = df["type"].value_counts()
count_of_M_Tv
```

```python
percentage = df["type"].value_counts(normalize=True) * 100
percentage
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

count_of_M_Tv.plot(kind="bar", edgecolor = "black" , linewidth = 3 , width= 0.4)



plt.title("Count of movies & Tv_shows" , fontsize = 15)
plt.xlabel("Category" , fontsize = 15)
plt.ylabel("Count" , fontsize = 15)


plt.xticks(rotation = 0)
plt.show
```

```python
rating_analysis = df["rating"].explode()
rating_analysis
```

```python
rating = rating_analysis.value_counts()
rating
```

```python
Top_Rating = rating_analysis.value_counts(ascending=False).head(10)
Top_Rating
```

```python

kids_rating = ['TV-Y', 'TV-Y7', 'G', 'PG']

kids_content = df[df["rating"].isin(kids_rating)]

kids_content.shape[0]
```

```python
percentage = (kids_content.shape[0] / df.shape[0]) * 100
print(f"Kids Content Percentage: {percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns


if not kids_content.empty:

    sns.countplot(
        y='rating',
        data=kids_content,
        order=kids_content['rating'].value_counts().index
    )

    plt.title("Kids Content Ratings")
    plt.xlabel("Count")
    plt.ylabel("Rating")


    plt.show()

else:
    print("No kids content found.")
```
